```python
import sys
import random
from datetime import datetime
import requests

def get_weather(city):
    api_key = "a73fbebc441e170a6a21dd2c2c82efae"  #API key
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        weather = data['weather'][0]['description']
        temp = data['main']['temp']
        print(f"The weather in {city.title()} is {weather} with a temperature of {temp}Â°C.")
    else:
        print("Sorry, I couldn't fetch the weather. Please check the city name or your connection.")



# Save tasks or reminders to a file and display them each time the assistant starts.
def save_reminder():
    with open("reminders.txt", "a") as file:
        reminder = input("Enter your reminder: ")
        file.write(reminder + "\n")
        print("Your reminder has been saved.")

def show_reminders():
    print("\nYour reminders:")
    try:
        with open("reminders.txt", "r") as file:
            reminders = file.readlines()
            for i, reminder in enumerate(reminders, 1):
                print(f"{i}. {reminder.strip()}")
    except FileNotFoundError:
        print("No reminders found.")


# Use an API like JokeAPI to fetch random jokes.
def get_joke():
    url = "https://v2.jokeapi.dev/joke/Any"
    response = requests.get(url)
    if response.status_code == 200:
        joke_data = response.json()
        if joke_data['type'] == 'single':
            print(f"\nHere's a joke: {joke_data['joke']}\n")
        else:
            print(f"Here's a joke: {joke_data['setup']}\n{joke_data['delivery']}")
    else:
        print("Sorry, I couldn't fetch a joke right now.")


# Let the user perform basic calculations (addition, subtraction, multiplication, division).
def calculator():
    print("Simple Calculator")
    num1 = float(input("Enter the first number: "))
    operation = input("Choose an operation (+, -, *, /): ")
    num2 = float(input("Enter the second number: "))
    
    if operation == "+":
        print(f"The result is: {num1 + num2}")
    elif operation == "-":
        print(f"The result is: {num1 - num2}")
    elif operation == "*":
        print(f"The result is: {num1 * num2}")
    elif operation == "/":
        if num2 != 0:
            print(f"The result is: {num1 / num2}")
        else:
            print("Error: Division by zero.")
    else:
        print("Invalid operation.")


# main menu system with options so the assistant runs until the user chooses to exit.
def personal_assistant():

    # grtting the user name 
    name = input("what is your name? ")
    print("Hello " + name.title() + '!' + " welcome to personal assistant.")

    # Use the datetime module to fetch and display the current date and time
    current_time = datetime.now()
    print("The current date and time is " + str(current_time))

    # Create a list of quotes.
    # Use the random module to pick and display one quote.
    quotes = [
        "Believe you can and you're halfway there.",
        "The only way to do great work is to love what you do.",
        "Success is not the key to happiness. Happiness is the key to success.",
        "Don't watch the clock; do what it does. Keep going.",
        "Act as if what you do makes a difference. It does."
    ]

    # choose the quote and display it for the user
    quote_of_the_day = random.choice(quotes)
    print(f"Here's your motivational quote: {quote_of_the_day}")

    while True:
        print("\n=== Personal Assistant Menu ===")
        print("1. Get Weather")
        print("2. Add Reminder")
        print("3. Show Reminders")
        print("4. Tell Me a Joke")
        print("5. Simple Calculator")
        print("6. Exit")
        
        choice = input("Choose an option (1-6): ")
        
        if choice == "1":
            city = input("Enter your city: ")
            get_weather(city)
        elif choice == "2":
            save_reminder()
        elif choice == "3":
            show_reminders()
        elif choice == "4":
            get_joke()
        elif choice == "5":
            calculator()
        elif choice == "6":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

# Start the assistant
personal_assistant()

sys.stdout.flush()


```
